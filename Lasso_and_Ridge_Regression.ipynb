**L2 Regularization** - Ridge regression

```python
import pandas as pd
import matplotlib.pyplot as pl
import numpy as np
import seaborn as sns
from matplotlib import rcParams
```

```python
df = pd.read_csv('/content/50_Startups.csv')
df.head(10)
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df.State = le.fit_transform(df.State)
```

```python
df
```

### **X and Y Split**

```python
Y = df['Profit']
X = df.drop(columns = ['Profit'],axis =1)
```

```python
X
```

### **Normalization**

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

X_scaled = pd.DataFrame(scaler.fit_transform(X), columns = X.columns)
```

### **Train Test Split**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, Y, test_size=0.2, random_state = 0)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import Lasso ,  Ridge
```

```python
l = Lasso()
r = Ridge()
```

```python
l.fit(X_train,Y_train)
```

```python
r.fit(X_train,Y_train)
```

```python
pred_r =r.predict(X_test)
pred_r
```

```python
pred_r_train = r.predict(X_train)
pred_r_train
```

```python
pred_l =l.predict(X_test)
pred_l
```

```python
pred_l_train = l.predict(X_train)
pred_l_train
```

```python
profit =pd.DataFrame({'Actual_Profit':Y_test, 'Ridge_pred':pred_r,'Lasso_pred':pred_l})
profit.head(10)
```

```python
from sklearn import metrics
```

```python
print('R2_Score')
print(f'Lasso : {metrics.r2_score(Y_test, pred_l)}')
print(f'Ridge : {metrics.r2_score(Y_test, pred_r)}')
```

```python
print('RMSE')
print(f'Lasso : {np.sqrt(metrics.mean_squared_error(Y_test, pred_l))}')
print(f'Ridge : {np.sqrt(metrics.mean_squared_error(Y_test, pred_r))}')

# Lesser RMSE is more better
```
